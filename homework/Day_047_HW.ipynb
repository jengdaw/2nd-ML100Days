了解如何使用 Sklearn 中的 hyper-parameter search 找出最佳的超參數

### 作業
請使用不同的資料集，並使用 hyper-parameter search 的方式，看能不能找出最佳的超參數組合

```python
from sklearn import datasets, metrics
from sklearn.model_selection import KFold, GridSearchCV, train_test_split
from sklearn.ensemble import GradientBoostingRegressor
```

```python
# 讀取資料集
digits = datasets.load_breast_cancer()

#切分訓練集/測試集
x_train, x_test, y_train, y_test = train_test_split(digits.data, digits.target, test_size=0.25, random_state=42)

# 建立模型
reg = GradientBoostingRegressor(random_state=7)
```

```python
# 先看看使用預設參數的結果
reg.fit(x_train, y_train)
y_pred = reg.predict(x_test)
print(metrics.mean_squared_error(y_test, y_pred))
```

```python
# 設定要訓練的超參數組合
n_estimators =[50, 100, 150]
max_depth = [1, 3, 5]
param_grid = dict( n_estimators=n_estimators, max_depth=max_depth)

#建立搜尋物件, 放入模型及參數組合字典
grid_search = GridSearchCV(reg, param_grid, scoring='neg_mean_squared_error', n_jobs=-1, verbose=1, cv=3)

#開始搜尋
grid_result =  grid_search.fit(x_train, y_train)
```

```python
#印出最佳結果與最佳參數
print("Best Accuracy : %f using %s" % (grid_result.best_score_, grid_result.best_params_))
```

```python
# 使用最佳參數建立最佳模型
 reg_bestparam = GradientBoostingRegressor(max_depth=grid_result.best_params_['max_depth'], n_estimators=grid_result.best_params_['n_estimators'])
# reg_bestparam = GradientBoostingRegressor(max_depth=1, n_estimators=150)

# 訓練模型
reg_bestparam.fit(x_train, y_train)
x_pred = reg_bestparam.predict(x_test)                                          
```

```python
#調整參數後
print(metrics.mean_squared_error(y_test, y_pred))
```
